```python
# Code to treat missing values
for i in df.columns:
  if df[i].dtype=='object': # datatype of column , categorical
    m=df[i].mode()[0] # modal value of categorical
    df[i]=df[i].fillna(m)
  else:
    me=df[i].mean() # mean value of every numerical column
    df[i]=df[i].fillna(me)

```
